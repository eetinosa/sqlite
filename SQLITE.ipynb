FORMAT RESULT

```python
cur.execute("SELECT * FROM auto")
items = cur.fetchall()

#print ("Age"+ "\t\t Sex"+"\t\t BMI"+"\t\t Children"+"\t\t Smoker"+"\t\t Region")
#print ("____________________________________________________________________________________________")

for item in items:
    age,sex,BMI,children,smoker,region=item
    print(f"{age}\t\t {sex}\t\t {BMI}\t\t{children}\t\t {smoker}\t\t {region}")
```

PRIMARY KEY ID

```python
cur.execute("SELECT rowid, * FROM auto")
items = cur.fetchall()

#print ("Age"+ "\t\t Sex"+"\t\t BMI"+"\t\t Children"+"\t\t Smoker"+"\t\t Region")
#print ("................................................................")

for item in items:
    pri_key,age,sex,BMI,children,smoker,region=item
    print(f"{pri_key}\t{age}\t\t {sex}\t\t {BMI}\t\t{children}\t\t{smoker}\t\t{region}")
```

WHERE CLAUSE

```python
cur.execute("SELECT * FROM auto WHERE age < '40'")
items = cur.fetchall()

print ("Age"+ "\t\t Sex"+"\t\t BMI"+"\t\t Children"+"\t Smoker"+"\t\t Region" + 
       "\n..........................................................................................")


for item in items:
    age,sex,BMI,children,smoker,region=item
    print(f"{age}\t\t {sex}\t\t {BMI}\t\t{children}\t\t{smoker}\t\t{region}")
```

# WHERE CLAUSE contd

```python
cur.execute("SELECT * FROM auto WHERE region LIKE 'north%'")
items = cur.fetchall()

print ("Age"+ "\t\t Sex"+"\t\t BMI"+"\t\t Children"+"\t Smoker"+"\t\t Region" + 
       "\n..........................................................................................")


for item in items:
    age,sex,BMI,children,smoker,region=item
    print(f"{age}\t\t {sex}\t\t {BMI}\t\t{children}\t\t{smoker}\t\t{region}")
```
